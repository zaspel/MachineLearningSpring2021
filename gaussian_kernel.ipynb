Load necessary modules

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.metrics import pairwise
```

Plot Gaussian kernel evaluated at different locations

```python
plt.figure(dpi=300)

X_eval = np.linspace(-3,3,100)
X_eval = X_eval.reshape(-1, 1)
x0 = np.zeros((1,1))
xm1 = -1*np.ones((1,1))
x1 = np.ones((1,1))
phi_X_eval = pairwise.rbf_kernel(X_eval, x0)
plt.plot(X_eval, phi_X_eval,label="$k(x,~~~0)$")
phi_X_eval = pairwise.rbf_kernel(X_eval, xm1)
plt.plot(X_eval, phi_X_eval,label="$k(x,-1)$")
phi_X_eval = pairwise.rbf_kernel(X_eval, x1)
plt.plot(X_eval, phi_X_eval,label="$k(x,~~~1)$")
plt.title("plot of Gaussian kernel $k(x,y)$ for different $y$ ($D=1$)")
plt.legend()
```

Plot Gaussian kernel evaluated for different kernel widths

```python
plt.figure(dpi=300)

X_eval = np.linspace(-3,3,100)
X_eval = X_eval.reshape(-1, 1)
x0 = np.zeros((1,1))
xm1 = -1*np.ones((1,1))
x1 = np.ones((1,1))

sigma_sq = 0.2
gamma = 1/sigma_sq
phi_X_eval = pairwise.rbf_kernel(X_eval, x0,gamma)
plt.plot(X_eval, phi_X_eval,label="$\sigma^2=0.2$")

sigma_sq = 1
gamma = 1/sigma_sq
phi_X_eval = pairwise.rbf_kernel(X_eval, x0,gamma)
plt.plot(X_eval, phi_X_eval,label="$\sigma^2=1$")

sigma_sq = 2
gamma = 1/sigma_sq
phi_X_eval = pairwise.rbf_kernel(X_eval, x0,gamma)
plt.plot(X_eval, phi_X_eval,label="$\sigma^2=5$")
plt.title("Plot of Gaussian kernel $k(x,0)$ for different $\sigma^2$ ($D=1$)")
plt.legend()
```
