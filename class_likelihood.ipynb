Load necessary modules

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

Generate artificial training data

```python
N = 30

np.random.seed(123)
X1 = np.random.multivariate_normal([1,1],[[1,0],[0,1]],N)
X1 = np.concatenate((X1,1.0*np.ones((N,1))),axis=1)
X2 = np.random.multivariate_normal([-1,-1],[[1,0],[0,1]],N)
X2 = np.concatenate((X2,2.0*np.ones((N,1))),axis=1)
X3 = np.random.multivariate_normal([2,0],[[1,0],[0,1]],N)
X3 = np.concatenate((X3,3.0*np.ones((N,1))),axis=1)

X = np.concatenate((X1,X2,X3));
```

Sketch likelihood of the three classes

```python
plt.figure()

cmap = matplotlib.cm.get_cmap('Pastel2')
val0=cmap(0.0)
val1=cmap(0.5)
val2=cmap(1.0)

plt.scatter(np.asarray([1,-1,2]),np.asarray([1,-1,0]), marker="o",s=500,color=val0)
plt.scatter(np.asarray([-1]),np.asarray([-1]), marker="o",s=500,color=val1)
plt.scatter(np.asarray([2]),np.asarray([0]), marker="o",s=500,color=val2)

plt.scatter(X[:,0],X[:,1],c=X[:,2],cmap='Pastel2', s=100);

ax = plt.gca()
circle1 = plt.Circle((1, 1), 2.4, color=val0, alpha=0.4)#, fill=trie)
circle2 = plt.Circle((-1, -1), 2.4, color=val1, alpha=0.4)#, fill=trie)
circle3 = plt.Circle((2, 0), 2.4, color=val2, alpha=0.4)#, fill=trie)
ax.add_artist(circle1)
ax.add_artist(circle2)
ax.add_artist(circle3)
ax.set_xlim((-3, 4))
ax.set_ylim((-3, 4))
plt.show()
```
