Load NumPy, Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

Generate and plot data

```python
np.random.seed(100)
N=100
mean = [0, 0]
cov = [[1, 0.8], [0.8, 1]]  # diagonal covariance
x, y = np.random.multivariate_normal([0,0], [[1,0.8],[0.8,1]], N).T
x = x-np.sum(x)/N
y = y-np.sum(y)/N
```

Compute principle components

```python
[U,D,Vt]=np.linalg.svd(np.vstack((x,y)).T,compute_uv=True)
V=Vt.T
```

Generate lines representing direction of first and second principle comonent

```python
samples = np.linspace(-0.25,0.25,400)
pc0 = D[0]*samples.reshape(-1,1)*V[:,0]
pc1 = D[1]*samples.reshape(-1,1)*V[:,1]
```

Plot everything

```python
plt.plot(x, y, '.')
plt.plot(pc0[:,0],pc0[:,1],c='orange', label="first principle component")
plt.plot(pc1[:,0],pc1[:,1],c='grey', label='second principle component')
plt.axis('equal')
plt.legend()
plt.show()
```
