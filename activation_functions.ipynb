Import necessary modules

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

Plot (scaled) commonly used activation functions

```python
plt.figure(dpi=300)
X_eval = np.linspace(-4,4,100)
X_eval = X_eval.reshape(-1, 1)
sigma = 1.0/(1.0+np.exp(-X_eval*5))
plt.plot(X_eval, sigma,label="sigmoid (scaled)")
sigma = 0.0+(X_eval>0)*X_eval
plt.plot(X_eval, sigma,label="ReLU")
sigma = np.log(1.0+np.exp(X_eval))
plt.plot(X_eval, sigma,label="softplus")
sigma = np.tanh(X_eval)
plt.plot(X_eval, sigma,label="tanh")
plt.ylim(-1.25,1.75)
plt.xticks([0],[0])
plt.xlabel("$v$")
plt.ylabel("$\sigma(v)$")
plt.title("commonly used activation functions")
plt.legend()
```
